```python
import pandas as pd
import math
import numpy as np

pd_file = pd.read_csv("16P.csv",encoding="ISO-8859-1") #data is readed as pandas frame
pd_file = pd_file.drop(columns=["Response Id"]) #response id is dropped
general_dict = dict()
general_dict["ESTJ"] = 0;general_dict["ENTJ"] = 1;general_dict["ESFJ"] = 2;general_dict["ENFJ"] = 3;general_dict["ISTJ"] = 4;general_dict["ISFJ"] = 5;general_dict["INTJ"] = 6;general_dict["INFJ"] = 7;general_dict["ESTP"] = 8
general_dict["ESFP"] = 9;general_dict["ENTP"] = 10;general_dict["ENFP"] = 11;general_dict["ISTP"] = 12;general_dict["ISFP"] = 13;general_dict["INTP"] = 14;general_dict["INFP"] = 15

pd_file["Personality"].replace(general_dict,inplace=True) 
list_of_names = pd_file.columns #it will be used for choosing column names.

persons_for_training = np.array(pd_file.iloc[:48000,:]) #first 48k data has been selected
#48000 is for person who we would like to train machine with that

persons_for_predicting = np.array(pd_file.iloc[48000:,:-1]) #12k data has been selected to predict.
persons_for_predicting_results = np.array(pd_file.iloc[48000:,-1]) #predicted datas result has been selected

all_misclassified = dict() #this will be used for detecting how many error occured
```

```python
def calculating_and_printingAccuracy_recall_precision_values(results,all_predict_list): #this func will be used for printing accuracy precall and precison values.
    allTP_list =0 #for detecting how many TP occurs.
    for a in general_dict: #for every personal ID
        TP = 0; TN = 0; FP = 0 ; FN = 0;result_counter = 0
        for b in results[:len(all_predict_list)]: 
            if b == general_dict[a]: #if from results b is equal to that personal ID:
                if b == all_predict_list[result_counter]: #if b and our predicted result is equal:
                    TP +=1
                else:
                    FN +=1 #else FN

            elif all_predict_list[result_counter] == general_dict[a]: #if our result has come but it is not its turn:
                FP +=1
                
            result_counter +=1
        allTP_list = allTP_list + TP #for collecting total TP number
        print("Precision value for "+a+" %"+str(100 * TP/(TP + FP)))
        print("Recall value for "+a+" %"+str(100 * TP/(TP + FN)))
    print("Accuracy is : %"+str(100* allTP_list/len(all_predict_list)))
```

```python
print("k for 1:")
counter = 0
all_predict_list = [] ; deleted_list = []
for a in persons_for_predicting:  # for our total12k data to predict:
    all_result_list = []
    for b in persons_for_training: #for every train point
        k = b[:-1] - a #for every row take out of that our predict row
        all_result_list.append(k.dot(k)) #that result been dot producted by itself to calculate euclian distance square and that element appended to result list
    

    asg1 = all_result_list.index(min(all_result_list)) #asg1 is equal to min element of all_list index.
    
    all_predict_list.append(persons_for_training[asg1][-1]) #that index is selected from training array.
    
    counter +=1
    
calculating_and_printingAccuracy_recall_precision_values(all_predict_list,all_predict_list) #func is used for taking results.
```

```python
def calculating_max_count_of_elements_index(x): #it calculates of the list x which index is the most counted index.
    max_index_list = []
    for a in x:
        max_index_list.append(x.count(a))

    checking_list = []
    for b in max_index_list:
        if b >2:
            return max_index_list.index(b)
        elif b >1:
            checking_list.append(max_index_list.index(b))
    if len(checking_list) == 4 or len(checking_list) ==0:
        return False
    else:
        return checking_list[0]
```

```python
def calculating_and_printingAccuracy_recall_precision_values_for_feature_normalization_output(results,all_predict_list):
    allTP_list =0
    for a in general_dict:
        TP = 0; TN = 0; FP = 0 ; FN = 0;result_counter = 0
        for b in results[:len(all_predict_list)]:

            if b == general_dict[a]:
                if b == all_predict_list[result_counter]:
                    TP +=1
                else:
                    FN +=1

            elif all_predict_list[result_counter] == general_dict[a]:
                FP +=1

            result_counter +=1
        allTP_list = allTP_list + TP
        print("With Feature Normalization Precision value for "+a+" %"+str(100 * TP/(TP + FP)))
        print("With Feature Normalization Recall value for "+a+" %"+str(100 * TP/(TP + FN)))
    print("With Feature Normalization Accuracy is : %"+str(100* allTP_list/len(all_predict_list)))
```
